```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame with columns 'CREDIT_SCORE' and 'DELINQUENT'

df = pd.read_csv('loan_level_500k.csv')

# Create a histogram for CREDIT_SCORE
plt.figure(figsize=(10, 6))
plt.hist(df['CREDIT_SCORE'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of CREDIT_SCORE')
plt.xlabel('CREDIT_SCORE')
plt.ylabel('Frequency')
plt.grid(True)

# Display the histogram
plt.show()

# Create a bar chart for DELINQUENT
plt.figure(figsize=(6, 4))
df['DELINQUENT'].value_counts().plot(kind='bar', color='salmon')
plt.title('Distribution of DELINQUENT')
plt.xlabel('DELINQUENT')
plt.ylabel('Count')
plt.xticks(rotation=0)  # Rotate x-axis labels if needed

# Display the bar chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame with columns 'CREDIT_SCORE' and 'DELINQUENT'
df = pd.read_csv('loan_level_500k.csv')

# Define the lower and upper limits for the y-axis
y_min = 350
y_max = 800

# Create a box plot within the specified y-axis limits
plt.figure(figsize=(8, 6))
plt.boxplot([df[(df['DELINQUENT'] == 0) & (df['CREDIT_SCORE'] >= y_min) & (df['CREDIT_SCORE'] <= y_max)]['CREDIT_SCORE'],
             df[(df['DELINQUENT'] == 1) & (df['CREDIT_SCORE'] >= y_min) & (df['CREDIT_SCORE'] <= y_max)]['CREDIT_SCORE']],
            labels=['Non-Delinquent', 'Delinquent'],
            patch_artist=True,
            boxprops=dict(facecolor='skyblue', color='black'),
            medianprops=dict(color='black'))
plt.title('CREDIT_SCORE Distribution by Delinquency Status')
plt.ylabel('CREDIT_SCORE')
plt.ylim(y_min, y_max)  # Set the y-axis limits
plt.show()
```
